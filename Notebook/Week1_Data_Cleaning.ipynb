```python
import pandas as pd

df = pd.read_excel("Retail Sales Dataset.xlsx")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

df[['Order Date', 'Ship Date']].head()
```

```python
(df['Ship Date'] < df['Order Date']).sum()
```

```python
df['Date_Issue'] = df['Ship Date'] < df['Order Date']
df['Date_Issue'].value_counts()
```

```python
df['Profit Margin %'] = round((df['Profit'] / df['Sales']) * 100, 2)
df['Profit Margin %'].head()
```

```python
df['Month'] = df['Order Date'].dt.month
df['Quarter'] = df['Order Date'].dt.quarter
df['Year'] = df['Order Date'].dt.year

df[['Order Date', 'Month', 'Quarter', 'Year']].head()
```

```python
df['Days to Ship'] = (df['Ship Date'] - df['Order Date']).dt.days

df['Days to Ship'].head()
```

```python
df[['Profit Margin %', 'Month', 'Quarter', 'Year', 'Days to Ship']].describe()
```

```python
df.to_csv("Retail_Sales_Cleaned.csv", index=False)
print("File saved successfully!")
```
